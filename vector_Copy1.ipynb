```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import skfuzzy as fuzz
```

```python
%matplotlib notebook

plt.rcParams['figure.figsize']=(6,6)
plt.rcParams['figure.dpi']=100

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(0, 0, 0, c='k', s=10) 

v1=[1,9,6]
v2=[7,8,3]
v3=[3,5,4]

ax.scatter(v1[0], v1[1], v1[2], c='r', s=50, label='vector 1', edgecolors='none') 
ax.scatter(v2[0], v2[1], v2[2], c='g', s=50, label='vector 2', edgecolors='none') 
ax.scatter(v3[0], v3[1], v3[2], c='b', s=50, label='vector 3', edgecolors='none')

#xs, ys, zs = np.meshgrid(np.arange(-10, 10, 1),
                      #np.arange(-10, 10, 1),
                      #np.arange(-10, 10, 1))


ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')

#computing directions
#vector1
v1_mag=np.sqrt(v1[0]**2 + v1[1]**2 + v1[2]**2)
v1_xdirect = (v1[0]/v1_mag)
v1_ydirect = (v1[1]/v1_mag)
v1_zdirect = (v1[2]/v1_mag)


#vector2
v2_mag=np.sqrt(v2[0]**2 + v2[1]**2 + v2[2]**2)
v2_xdirect = (v2[0]/v2_mag)
v2_ydirect = (v2[1]/v2_mag)
v2_zdirect = (v2[2]/v2_mag)


#vector1
v3_mag=np.sqrt(v3[0]**2 + v3[1]**2 + v3[2]**2)
v3_xdirect = (v3[0]/v3_mag)
v3_ydirect = (v3[1]/v3_mag)
v3_zdirect = (v3[2]/v3_mag)

#ax.axis(xmin=0, xmax=100)
#ax.axis(ymin=0, ymax=100)
#ax.axis(zmin=0, zmax=100)
#ax.scatter(0, 0, 0, c='k', s=10) 
ax.quiver(0, 0, 0, v1_xdirect, v1_ydirect, v1_zdirect, length = v1_mag, normalize = True, color='r')
ax.quiver(0, 0, 0, v2_xdirect, v2_ydirect, v2_zdirect, length = v2_mag, normalize = True, color='g')
ax.quiver(0, 0, 0, v3_xdirect, v3_ydirect, v3_zdirect, length = v3_mag, normalize = True, color='b')

#plt.quiver(0, 0, 0, 1, 1, 1, scale=1, color='g')
ax.legend()
plt.show()
```
